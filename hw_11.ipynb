```python
import os
import pprint
import shutil

def ShowCurFileDir(dir_name = None):                                   #Функция вывода текущего каталога и файлов в нем
    if not dir_name :                                                  #Если путь не задан
        print("Текущий каталог :"+os.getcwd())                         #Выводим полный путь текущего каталога
        print("Содержит в себе следующие файлы и каталоги:")
        pprint.pprint(os.listdir())                                    #Выводим содержимое текущего каталога        
    elif os.path.exists(dir_name):                                     #Если путь задан и он верный
        print("Заданный путь каталога: "+dir_name)                     #Выводим полный путь заданного каталога
        print("Содержит в себе следующие файлы и каталоги:")
        pprint.pprint(os.listdir(dir_name))                            #Выводим содержимое заданного каталога        
    else:                                                              #Если задан неверный путь выводим ошибку
        print("Не верно задан путь каталога или такого каталога не существует")

def CreateDir(dir_name, dir_path = None):                              #Функция создания каталога
    if not dir_path:                                                   #Если создаем каталог в текущем (т.е. не задан путь)
        if not os.path.exists(dir_name):                               #Проверяем что каталога с таким же именем нет
            os.mkdir(dir_name)                                         #Создаем каталог
            print(f'каталог {dir_name} успешно создан')           
    elif os.path.exists(dir_path):                                     #Создаем каталог по заданному пути
        if not os.path.exists(os.path.join(dir_path, dir_name)):       #Проверяем что каталога с таким же именем нет
            os.mkdir(os.path.join(dir_path, dir_name))                 #Создаем каталог
            print(f'каталог {dir_name} успешно создан в каталоге {dir_path}')          
    else:                                                              #Если не верно задан путь
        print(f'Не верно задан исходный каталог {dir_path} или он не существует')

def CreateFile(file_name, dir_path = None):                            #Функция создания Файла
    if not dir_path:                                                   #Если создаем файл в текущем каталоге
        if not os.path.exists(file_name):                              #Проверяем что файла с таким же именем нет
            with open(file_name, "w") as f:                            #Создаем файл
                f.write("")
            print(f'Файл {file_name} успешно создан') 
    elif os.path.exists(dir_path):                                     #Создаем каталог по заданному пути
        if not os.path.exists(os.path.join(dir_path, file_name)):      #Проверяем что каталога с таким же именем нет
            with open(os.path.join(dir_path, file_name), "w") as f:    #Создаем файл
                f.write("")
            print(f'Файл {file_name} успешно создан в каталоге {dir_path}')  
    else:                                                              #Если не верно задан путь
        print(f'Не верно задан исходный каталог {dir_path} или он не существует')

def DeleteFile(file_name, dir_path = None):                            #Функция удаления файла
    if not dir_path:                                                   #Если удаляем файл в текущем каталоге
        if os.path.exists(os.path.join(os.getcwd(), file_name)):       #Проверяем что файл с заданным именем есть
            os.remove(file_name)
            print(f'Файл {file_name} успешно удален') 
    elif os.path.exists(dir_path):                                     #Если удаляем файл в заданном каталоге
        if os.path.exists(os.path.join(dir_path, file_name)):          #Проверяем что файл с заданным именем есть
            os.remove(os.path.join(dir_path, file_name))               #Удаляем файл по заданному пути
            print(f'Файл {file_name} успешно удален из каталога {dir_path}')
    else:                                                              #Если не верно задан путь
        print(f'Не верно задан исходный каталог {dir_path} или он не существует')

def DeleteDir(dir_name, dir_path = None):                              #Функция удаления каталога
    if not dir_path:                                                   #Если удаляем каталог из текущего
        if os.path.exists(os.path.join(os.getcwd(), dir_name)):        #Проверяем что имя каталога задано верно
            os.rmdir(dir_name)
            print(f'каталог {dir_name} успешно удален') 
    elif os.path.exists(dir_path):                                     #Если удаляем каталог по заданному пути
        if os.path.exists(os.path.join(dir_path, dir_name)):           #Проверяем что каталог по заданному пути существует
            os.rmdir(os.path.join(dir_path, dir_name))                 #Удаляем каталог
            print(f'каталог {dir_name} успешно удален из {dir_path}')
    else:                                                              #Если не верно задан путь
        print(f'Не верно задан исходный каталог {dir_path} или он не существует')

def DeleteFullDir(dir_name, dir_path = None):                          #Функция удаления всего каталога
    if not dir_path:                                                   #Если удаляем каталог из текущего
        if os.path.exists(os.path.join(os.getcwd(), dir_name)):        #Проверяем что имя каталога задано верно
            shutil.rmtree(dir_name)
            print(f'каталог {dir_name} с вложением успешно удален') 
    elif os.path.exists(dir_path):                                     #Если удаляем каталог по заданному пути
        if os.path.exists(os.path.join(dir_path, dir_name)):           #Проверяем что каталог по заданному пути существует
            shutil.rmtree(os.path.join(dir_path, dir_name))            #Удаляем каталог
            print(f'каталог {dir_name} с вложением успешно удален из {dir_path}')
    else:                                                              #Если не верно задан путь
        print(f'Не верно задан исходный каталог {dir_path} или он не существует')

def rename_file(old_file_name, new_file_name):                         #Функция для переименования файла или каталога
    if os.path.exists(old_file_name):                                  #Проверяем на правильность задания имени файла
        os.rename(old_file_name, new_file_name)
        print("Файл успешно переименован.")
    else:
        print("Файл не существует.")

def copy_file(file_name, destination):                                 #Функция для копирования файла
    if os.path.exists(file_name):
        shutil.copy(file_name, destination)
        print("Файл успешно скопирован.")
    else:
        print("Файл не существует.")

def move_file(file_name, destination):                                 #Функция для перемещения файла
    if os.path.exists(file_name):
        shutil.move(file_name, destination)
        print("Файл успешно перемещен.")
    else:
        print("Файл не существует.")
from IPython.display import clear_output


def main():
    while True:
        print('Работа с файлами:')
        print('1. Просмотреть содержимое каталога')
        print('2. Создать каталог')
        print('3. Создать файл')
        print('4. Удалить каталог')
        print('5. Удалить файл')
        print('6. Удалить каталог полностью')
        print('7. Переименовать файл или каталог')
        print('8. Скопировать файл или каталог')
        print('9. Переместить файл или каталог')
        print('0. Выход')
        print()
        try:
            vybor = int(input("Выберите пункт меню: "))
        except ValueError:
            print('Ошибка ввода. Попробуйте еще раз')    
        
        match vybor:
            case 1:                                                    #Просмотреть содержимое каталога
                dir_name = input('Введите полный путь каталога или оставьте поле пустым для отображения текущего каталога:\n')
                ShowCurFileDir(dir_name)
                print()
            case 2:                                                    #Создать каталог
                print('Задайте имя создаваемого каталога, а затем задате путь для его создания/n')
                dir_name = input('Введите название нового каталога: ')
                dir_path = input('Введите полный путь или оставьте пустым для создания в текущей директории: ')
                CreateDir(dir_name, dir_path)
                print()
            case 3:                                                    #Создать файл
                print('Задайте имя создаваемого файла, а затем задате путь для его создания/n')
                file_name = input('Введите название нового файла: ')
                dir_path = input('Введите полный путь или оставьте пустым для создания в текущей директории: ')
                CreateFile(file_name, dir_path)
                print()
            case 4:                                                    #Удалить каталог
                dir_path = input('Введите путь до каталога или оставьте пустым для удаления в текущей директории: ')
                dir_name = input('Введите название каталога для удаления: ')
                DeleteDir(dir_name, dir_path)
                print()
            case 5:                                                    #Удалить файл
                dir_path = input('Введите полный путь до удаляемого файла или оставьте пустым для удаления файла в текущей директории: ')
                file_name = input('Введите название файла для удаления: ')
                DeleteFile(file_name, dir_path)
                print()
            case 6:                                                    #Удалить каталог полностью
                dir_path = input('Введите путь до каталога или оставьте пустым для удаления в текущей директории: ')
                dir_name = input('Введите название каталога для удаления: ')
                DeleteFullDir(dir_name, dir_path)
                print()
            case 7:                                                    #Переименовать файл или каталог
                print('Задайте имя для переименования файла (можно указывать полный путь до файла)')
                old_file_name = input('Имя файла для переименования (текущий каталог можно задать через ./): ')
                new_file_name = input('Введите новое имя файла (абсолютный путь текщего каталога задается как ./): ')
                rename_file(old_file_name, new_file_name)
                print() 
            case 8:                                                    #Скопировать файл или каталог
                file_name = input('Введите имя файла или папки для копирования:')
                destination = input('Введите путь назначения для копирования (абсолютный путь текщего каталога задается как ./): ')
                copy_file(file_name, destination)
                print()              
            case 9:                                                    #Переместить файл или каталог
                file_name = input('Введите имя файла или папки для перемещения:')
                destination = input('Введите путь назначения для перемещения: ')
                move_file(file_name, destination)
                print()
            case 0:                                                    #Выход
                print('До новых встреч!')
                break                
            case _:
                clear_output()                                         #Очищаем консоль
                print('Ошибка. Выберите корретный пункт. ') 

if __name__=='__main__':
    main()
```
